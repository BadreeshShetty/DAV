```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas_datareader as pdr
from datetime import datetime
sns.set_style('whitegrid')
```

```python
# Set up End and Start times for data grab
end = datetime.now()
start = datetime(end.year - 1,end.month,end.day)

data_aapl = pdr.get_data_yahoo('AAPL',start,end)
data_aapl.head()
```

```python
data_goog = pdr.get_data_yahoo('GOOG',start,end)
data_goog.head()
```

```python
data_msft = pdr.get_data_yahoo('MSFT',start,end)
data_msft.head()
```

```python
data_amzn = pdr.get_data_yahoo('AMZN',start,end)
data_amzn.head()
```

```python
aapl_adj=pd.DataFrame(data_aapl['Adj Close'])
goog_adj=pd.DataFrame(data_goog['Adj Close'])
msft_adj=pd.DataFrame(data_msft['Adj Close'])
amzn_adj=pd.DataFrame(data_amzn['Adj Close'])
```

```python
plt.figure(figsize=(15,10));

plt.plot(data_aapl['Adj Close']);
plt.plot(data_goog['Adj Close']);
plt.plot(data_msft['Adj Close']);
plt.plot(data_amzn['Adj Close']);
plt.legend(["AAPL", "GOOG","MSFT","AMZN"]);
plt.title("Adj Close");
```

```python
aapl_adj['MA_10'] = aapl_adj['Adj Close'].rolling(10).mean().shift()
aapl_adj['MA_20'] = aapl_adj['Adj Close'].rolling(20).mean()
aapl_adj['MA_30'] = aapl_adj['Adj Close'].rolling(30).mean()
plt.figure(figsize=(15,10))
plt.plot(aapl_adj['Adj Close'],label='AAPL')
plt.plot(aapl_adj['MA_10'], label='MA 10 day')
plt.plot(aapl_adj['MA_20'], label='MA 20 day')
plt.legend(loc=2);
```

```python
goog_adj['MA_10'] = goog_adj['Adj Close'].rolling(10).mean().shift()
goog_adj['MA_20'] = goog_adj['Adj Close'].rolling(20).mean()
goog_adj['MA_30'] = goog_adj['Adj Close'].rolling(30).mean()

plt.figure(figsize=(15,10))
plt.plot(goog_adj['Adj Close'],label='GOOG')
plt.plot(goog_adj['MA_10'], label='MA 10 day')
plt.plot(goog_adj['MA_20'], label='MA 20 day')
plt.legend(loc=2);
```

```python
msft_adj['MA_10'] = msft_adj['Adj Close'].rolling(10).mean().shift()
msft_adj['MA_20'] = msft_adj['Adj Close'].rolling(20).mean()
msft_adj['MA_30'] = msft_adj['Adj Close'].rolling(30).mean()

plt.figure(figsize=(15,10))
plt.plot(msft_adj['Adj Close'],label='MSFT')
plt.plot(msft_adj['MA_10'], label='MA 10 day')
plt.plot(msft_adj['MA_20'], label='MA 20 day')
plt.legend(loc=2);
```

```python
amzn_adj['MA_10'] = amzn_adj['Adj Close'].rolling(10).mean().shift()
amzn_adj['MA_20'] = amzn_adj['Adj Close'].rolling(20).mean()
amzn_adj['MA_30'] = amzn_adj['Adj Close'].rolling(30).mean()

plt.figure(figsize=(15,10))
plt.plot(amzn_adj['Adj Close'],label='AMZN')
plt.plot(amzn_adj['MA_10'], label='MA 10 day')
plt.plot(amzn_adj['MA_20'], label='MA 20 day')
plt.legend(loc=2);
```

```python
aapl_adj['Daily Return'] = aapl_adj['Adj Close'].pct_change()
aapl_adj['Daily Return'].plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
plt.title("AAPL Daily Return");
```

```python
goog_adj['Daily Return'] = goog_adj['Adj Close'].pct_change()
goog_adj['Daily Return'].plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
plt.title("GOOG Daily Return");
```

```python
msft_adj['Daily Return'] = msft_adj['Adj Close'].pct_change()
msft_adj['Daily Return'].plot(figsize=(12,4),legend=True,linestyle='--',marker='o')
plt.title("MSFT Daily Return");
```
